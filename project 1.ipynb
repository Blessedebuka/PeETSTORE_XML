```python
# IO GET THE INDIVIDUAL VALUE OF COUNTRIES TAKING PYTHON
```

```python
python_INDIVI=df.groupby(['Country'])
```

```python
Python_Users=python_INDIVI['LanguageWorkedWith'].apply(lambda x:x.str.contains('Python').sum())# TO GET THE SUM OF INDIVIDUAL COUNTRIES HAVING PYTHON
```

```python
Python_Users
```

```python
Country_Respondents=df['Country'].value_counts()  # NO OF RESPONDENTS IN EACH COUNTRY
```

```python
Country_Respondents
```

```python
Python_df = pd.concat([Python_Users,Country_Respondents],axis='columns', sort=False) # TO TO INCLUDE 'Python_Users' AND 'Country_Respondents' IN A DATAFRAME
```

```python
Python_df
```

```python
Python_df.rename(columns={'LanguageWorkedWith':'NumKnowsPython','Country':'NumOfRespondents'},inplace =True)
Python_df # TO RENAME COLUMNS 'Country' AND 'LanguageWorkedWith' WITH 'NumOfRespondents' and 'NumKnowsPython' RESPECTIVELY IN THE Python_df dataframe
```

```python
Python_df['PercentKnowsPython']= (Python_df['NumKnowsPython']/ Python_df['NumOfRespondents']) * 100
Python_df['PercentKnowsPython']
Python_df  #TO CALCULATE THE % THAT KNOWS PYTHON AND INCLUDE IT WITH NO THAT KNOWS PYTHON AND NO OF RESPONDENTS IN A DATAFRAME
```

```python
Python_df.sort_values(by='PercentKnowsPython', ascending =False,inplace=True) # TO SORT THE Python_df IN DESCENDING ORDER
Python_df
```
